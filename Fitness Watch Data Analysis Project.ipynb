# Analysis Data

This code plots histograms for the numerical columns in a DataFrame using the matplotlib library. It displays 20 bins for each distribution with the plots arranged in a 5-row by 3-column grid. The goal is to analyze and visualize the data distribution to understand value patterns and detect any trends.

```python
plt.figure(figsize=(15, 12))

# Plotting the histograms for all numerical columns
df.hist(bins=20, figsize=(15, 12), layout=(5, 3))

plt.tight_layout()
plt.show()
```

# Frequency Distribution of Days of the Week

```python
# Frequency Distribution of Days of the Week

plt.figure(figsize = ((8,4)))
plt.hist(df.DayOfWeek , bins=7 , color = 'lightskyblue', width= 0.6 )


plt.xlabel('Day Of Week')
plt.ylabel('freq')

plt.title('freq by day of week')
plt.show()
```

# Calories vs Total Steps

```python
#Calories vs Total Steps
plt.figure(figsize=(8,6))
plt.scatter(df.TotalSteps , df.Calories , c= df.Calories)

median_Steps = 5986
median_Calories = 1440

plt.axhline(median_Calories , color = 'blue' , label= ' median of calories')
plt.axvline(median_Steps , color = 'red' , label= ' median of steps')

plt.xlabel('Steps')
plt.ylabel('Calories')

plt.title(' Calories By Total Steps')
plt.show()
```

# Calories vs Total Hours 

```python
# Calories vs Total Hours 
plt.figure(figsize=(8,6))
plt.scatter(df.TotalHours , df.Calories , c= df.Calories)

median_Hours = 24
median_Calories = 1440

plt.axhline(median_Calories , color = 'blue' , label= ' median of calories')
plt.axvline(median_Hours , color = 'red' , label= ' median of Hours')

plt.xlabel('Hours')
plt.ylabel('Calories')

plt.title(' Calories By Total Hours')
plt.show()
```

# Activity Minutes Distribution

```python
#Activity Minutes Distribution
VeryActiveMinutes = df.VeryActiveMinutes.sum()
FairlyActiveMinutes = df.FairlyActiveMinutes.sum()
LightlyActiveMinutes = df.LightlyActiveMinutes.sum()
SedentaryMinutes= df.SedentaryMinutes.sum()

Minutes = [VeryActiveMinutes ,	FairlyActiveMinutes ,	 LightlyActiveMinutes , 	SedentaryMinutes	]
labels = [ 'VeryActiveMinutes' ,	'FairlyActiveMinutes' ,	' LightlyActiveMinutes' , 'SedentaryMinutes'	]

plt.pie(Minutes , labels = labels , autopct='%1.1f%%' , explode = [0,0,0,0] )
```

1. Sedentary minutes have the biggest percentage at  83.3%.
2. This might indicate users are using the app for daily routine and while running errands
3. There is a  minor percentage of fairly active activity (1.1%) and very active activity (1.4%) which can be a cause of concern as the app is made to promote fitness.

# Total Active Minutes by Day of the Week

```python
# Group data by 'DayOfWeek' and sum the activity minutes for each day
daily_activity = df.groupby('DayOfWeek')[['VeryActiveMinutes', 'FairlyActiveMinutes', 'LightlyActiveMinutes']].sum()


daily_activity.plot(kind='bar', figsize=(10, 6))


plt.xlabel('Day of the Week')
plt.ylabel('Minutes')
plt.title('Active Minutes for Each Day of the Week')


plt.xticks(rotation=0) 
plt.show()
```

# Conclusions and Recommendations

# Conclusions:
Based on the analysis of FitBit data, the following key insights were observed:

1. Daily App Usage:
- Users engage more with the FitBit app during weekdays than on weekends, likely due to routine and work commitments.
- A noticeable decrease in app usage over weekends suggests reduced tracking of physical activities during this time.
2. Physical Activity:

- Higher levels of physical activity, including steps and calories burned, are observed during weekdays.
- There are significant variations in activity levels across users, with some being very active and others sedentary.

3. Sedentary Behavior:

- Extended periods of sedentary behavior are common, especially on weekdays, possibly linked to office work or studying.
- Sedentary behavior increases on weekends, indicating a preference for relaxation and non-physical activities.
# Recommendations:
To boost user engagement and improve marketing strategies, we suggest:

1. Daily Reminders:

 - Send daily notifications encouraging users to track activities, reducing sedentary behavior.
2. Weekend Challenges:

- Implement weekend activity challenges to keep users engaged and active during weekends, such as step goals or calorie-burning targets.
3. Rewards and Badges:

- Offer incentives like badges or discounts for users who meet health goals, enhancing motivation.
4. User Group Segmentation:

- Segment users based on activity levels and tailor content accordingly, offering intensive workout tips for active users and motivational - content for less active users.
5. UI Improvements:

- Enhance the app’s user interface, adding features like heat maps and performance analysis tools to simplify tracking of both physical and sedentary activities.
# Summary:
The analysis of FitBit data provides key insights into user behavior and activity patterns, which can guide the development of effective marketing strategies. By implementing these recommendations, Bellabeat can enhance user engagement, improve user experience, and strengthen its position as a leader in health and wellness technology.
